```python
brand_avg = j3.groupby("coupon_id")['brand'].mean().to_dict()
j2['brand_avg'] = j2['coupon_id'].map(brand_avg)

brand_type_avg = j3.groupby("coupon_id")['brand_type'].mean().to_dict()
j2['brand_type_avg'] = j2['coupon_id'].map(brand_type_avg)

category_avg = j3.groupby("coupon_id")['category'].mean().to_dict()
j2['category_avg'] = j2['coupon_id'].map(category_avg)


brand_avg = j3.groupby("coupon_id")['brand'].max().to_dict()
j2['brand_max'] = j2['coupon_id'].map(brand_avg)

brand_type_avg = j3.groupby("coupon_id")['brand_type'].max().to_dict()
j2['brand_type_max'] = j2['coupon_id'].map(brand_type_avg)

category_avg = j3.groupby("coupon_id")['category'].max().to_dict()
j2['category_max'] = j2['coupon_id'].map(category_avg)


brand_avg = j3.groupby("coupon_id")['brand'].min().to_dict()
j2['brand_min'] = j2['coupon_id'].map(brand_avg)

brand_type_avg = j3.groupby("coupon_id")['brand_type'].min().to_dict()
j2['brand_type_min'] = j2['coupon_id'].map(brand_type_avg)

category_avg = j3.groupby("coupon_id")['category'].min().to_dict()
j2['category_min'] = j2['coupon_id'].map(category_avg)


brand_avg = j3.groupby("coupon_id")['brand'].nunique().to_dict()
j2['brand_count'] = j2['coupon_id'].map(brand_avg)

brand_type_avg = j3.groupby("coupon_id")['brand_type'].nunique().to_dict()
j2['brand_type_count'] = j2['coupon_id'].map(brand_type_avg)

category_avg = j3.groupby("coupon_id")['category'].nunique().to_dict()
j2['category_count'] = j2['coupon_id'].map(category_avg)
```
